```python
df_articles = pd.DataFrame(columns=['year', 'month', 'title', 'text'])

path_nplus1 = os.path.join(path, 'external', 'nplus1')
for f_year in list_nplus1:
    try:
        print(f'year: {f_year}')

        for f_month in os.listdir(os.path.join(path_nplus1, f_year)):
            #print(f'\tmonth: {f_month}')

            path_nplus1_year_month = os.path.join(path_nplus1, f_year, f_month)
            print(f'\t month: {f_month} | shape articles: {len(path_nplus1_year_month)}')
            for f_article in os.listdir(path_nplus1_year_month):
                take_normal_name_title = ' '.join(f_article.split('.')[0].split('_'))

                with open(f'{path_nplus1_year_month}/{f_article}', 'r') as f:
                    text_article = f.read()

                df_articles = df_articles.append({'year': f_year,
                                                  'month': f_month,
                                                  'title': take_normal_name_title,
                                                  'text': text_article},
                                                 ignore_index=True
                                                )
    except Exception as e:
        print(f'ERROR: {e}')
```

```python
df_articles['id'] = list(range(1,df_articles.shape[0]+1))
df_articles = df_articles[['id','year', 'month', 'title', 'text']]
print(df_articles.shape)
df_articles.head()
```
